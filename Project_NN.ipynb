```python
# Pandas is used for data manipulation
import pandas as pd
# Read in data and display first 5 rows
df = pd.read_csv('C:/Users/shelb/OneDrive/Documents/MSA2021/Machine Learning/train.csv')
valid=pd.read_csv('C:/Users/shelb/OneDrive/Documents/MSA2021/Machine Learning/valid.csv')
test=pd.read_csv('C:/Users/shelb/OneDrive/Documents/MSA2021/Machine Learning/test.csv')
df.head(5)
```

```python
print('The shape of our features is:', df.shape)
print('The shape of our features is:', valid.shape)
print('The shape of our features is:', test.shape)
```

```python
# Descriptive statistics for each column
df.describe()
```

```python
cleanup_nums = {"v45":     {'Monday':0, 'Tuesday':1, 'Wednesday':2, 'Thursday':3, 'Friday':4}}
df.replace(cleanup_nums, inplace=True)
valid.replace(cleanup_nums, inplace=True)
test.replace(cleanup_nums, inplace=True)
```

```python
# Use numpy to convert to arrays
import numpy as np
# Labels are the values we want to predict
y = np.array(df['cont.target'])
y_valid = np.array(valid['cont.target'])
# Remove the labels from the features
# axis 1 refers to the columns
x= df.drop(['cont.target','binary.target1','binary.target2','v7'], axis = 1)
x_valid= valid.drop(['cont.target','binary.target1','binary.target2','v7'], axis = 1)
x_test= test.drop(['v7'], axis = 1)
# Saving feature names for later use
#feature_list = list(features.columns)
# Convert to numpy array
x = np.array(x)
x_valid = np.array(x_valid)
x_test = np.array(x_test)
```

```python
print('Training Features Shape:', x.shape)
print('Training Labels Shape:', y.shape)
print('Testing Features Shape:', x_valid.shape)
print('Testing Labels Shape:', y_valid.shape)
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
# Train the model on training data
rf.fit(x, y);
```

```python
# Use the forest's predict method on the test data
predictions = rf.predict(x_valid)
# Calculate the absolute errors
errors = abs(predictions - y_valid)
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
Mean Absolute Error: 3.83 degrees.
```

```python
# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / y_valid)
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
Accuracy: 93.99 %.
```
